```python
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt

# Hybrid data
hybrid_threads = np.array([5, 5, 6, 7, 8, 9])
hybrid_ranks = np.array([2, 3, 2, 2, 2, 3])
hybrid_time = np.array([0.393592, 0.258655, 0.300122, 0.398182, 0.492336, 0.496613])

# Normalize time for color mapping
norm = plt.Normalize(hybrid_time.min(), hybrid_time.max())
colors = plt.cm.viridis(norm(hybrid_time))

# Create 3D figure
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot with color mapping based on execution time
sc = ax.scatter(hybrid_threads, hybrid_ranks, hybrid_time,
                c=hybrid_time, cmap='viridis', s=100, edgecolor='k', depthshade=True)

# Titles and labels
ax.set_title('Hybrid Parallel Execution Time (OMP + MPI)', fontsize=14, fontweight='bold')
ax.set_xlabel('OpenMP Threads', fontsize=12)
ax.set_ylabel('MPI Ranks', fontsize=12)
ax.set_zlabel('Max Execution Time (s)', fontsize=12)

# Optional: Customize ticks
ax.set_xticks(sorted(set(hybrid_threads)))
ax.set_yticks(sorted(set(hybrid_ranks)))

# Add colorbar
cbar = fig.colorbar(sc, ax=ax, shrink=0.5, aspect=10)
cbar.set_label('Execution Time (s)', fontsize=12)

# Improve layout
ax.view_init(elev=20, azim=135)  # Adjust viewing angle
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
searching_values = [1, 2, 3, 4, 5]

serial_time = [0.000224, 0.000293, 0.000512, 0.000820, 0.001013]
openmp_time = [0.090305, 0.104331, 0.142559, 0.227195, 0.312031]
mpi_time = [0.089665, 0.100138, 0.146799, 0.182519, 0.259694]

# Plot
plt.figure(figsize=(10, 6))
plt.plot(searching_values, serial_time, marker='o', label='Serial', linewidth=2)
plt.plot(searching_values, openmp_time, marker='s', label='OpenMP', linewidth=2)
plt.plot(searching_values, mpi_time, marker='^', label='MPI', linewidth=2)

# Labels and styling
plt.title('Execution Time Comparison: Serial vs OpenMP vs MPI', fontsize=14)
plt.xlabel('Number of Searching Values', fontsize=12)
plt.ylabel('Execution Time (s)', fontsize=12)
plt.xticks(searching_values)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(fontsize=11)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Data
search_vals = [1, 2, 3, 4, 5]
serial_time = np.array([0.000224, 0.000293, 0.000512, 0.000820, 0.001013])
openmp_time = np.array([0.090305, 0.104331, 0.142559, 0.227195, 0.312031])
mpi_time = np.array([0.089665, 0.100138, 0.146799, 0.182519, 0.259694])

# Calculate Speedup
openmp_speedup = serial_time / openmp_time
mpi_speedup = serial_time / mpi_time

# Create table
df = pd.DataFrame({
    'Number of Searching Values': search_vals,
    'Serial Time (s)': serial_time,
    'OpenMP Time (s)': openmp_time,
    'MPI Time (s)': mpi_time,
    'OpenMP Speedup': openmp_speedup,
    'MPI Speedup': mpi_speedup
})

# Display table
print("📊 Speedup Table (Serial vs OpenMP vs MPI):")
print(df.to_string(index=False))

# Plot Speedup Graph
plt.figure(figsize=(10, 6))
plt.plot(search_vals, openmp_speedup, marker='o', label='OpenMP Speedup', color='green', linewidth=2)
plt.plot(search_vals, mpi_speedup, marker='s', label='MPI Speedup', color='red', linewidth=2)

plt.title('Speedup over Serial: OpenMP vs MPI', fontsize=14)
plt.xlabel('Number of Searching Values', fontsize=12)
plt.ylabel('Speedup (Serial Time / Parallel Time)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.xticks(search_vals)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Data
search_vals = [5, 6, 7, 8, 9]
serial_time = np.array([0.001013, 0.001632, 0.004138, 0.005186, 0.008320])
hybrid_time = np.array([0.258655, 0.300122, 0.398182, 0.492336, 0.496613])

# Calculate Speedup
hybrid_speedup = serial_time / hybrid_time

# Create table
df = pd.DataFrame({
    'Number of Searching Values': search_vals,
    'Serial Time (s)': serial_time,
    'Hybrid Time (s)': hybrid_time,
    'Speedup': hybrid_speedup
})

# Display table
print("📊 Speedup Table (Serial vs Hybrid):")
print(df.to_string(index=False))

# Plot Speedup Graph
plt.figure(figsize=(10, 6))
plt.plot(search_vals, hybrid_speedup, marker='o', color='purple', linewidth=2, label='Hybrid Speedup')

plt.title('Speedup of Hybrid Method Over Serial', fontsize=14)
plt.xlabel('Number of Searching Values', fontsize=12)
plt.ylabel('Speedup (Serial Time / Hybrid Time)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.xticks(search_vals)
plt.axhline(1.0, color='gray', linestyle='--', label='No Speedup (x1)')
plt.legend()
plt.tight_layout()
plt.show()
```
